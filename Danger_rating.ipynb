```python
import os
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
n=pd.read_csv("neighbourhood.csv")
top10 = n.nlargest(100, 'Danger rating')
plt.figure(figsize=(20, 8))
plt.bar(top10['Neighbourhood'], top10['Danger rating'])
plt.xlabel('Neighborhood')
plt.ylabel('Safety Rating')
plt.title('Bottom 10 neighbourhoods according to our criteria')
plt.xticks(rotation='vertical')
plt.show()
```
