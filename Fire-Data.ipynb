## Fire Data for neighborhoods in Pittsburgh

The following is an analysis of data about the number of fire incidents in the neighborhoods in Pittsburgh. The metric used to evalute each neighborhood is the amount of fires, where a less amount of fires equates to a better neighborhood.|

```python
#imports pandas
import pandas as pd
import matplotlib.pyplot as plt

#creates a list of every neighborhood and its fire count in decending order
fireIncidents = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02")
pd.set_option('display.max_rows', None)
fireIncidents['neighborhood'].value_counts()
```

```python
#creates a bar graph of the 10 neighborhoods with the most amount of fires which represents the worst 10 neighborhoods
bottom_10 = (fireIncidents['neighborhood'].value_counts()).iloc[:10]
plt.subplots(figsize=(7, 4))
bottom_10.plot.bar(color = 'red')
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Bottom 10 Neighborhoods by Number of Fires')
plt.show()
```

```python
#creates a bar graph of the 10 neighborhoods with the least amount of fires which represents the best 10 neighborhoods
top_10 = (fireIncidents['neighborhood'].value_counts()).iloc[-10:]
plt.subplots(figsize=(7, 4))
top_10.plot.bar(color = 'green').invert_xaxis()
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Top 10 Neighborhoods by Number of Fires')
plt.show()
```

```python
#creates a bar graph of all of the neighborhoods to help visualize the difference in the number of fires between each
labels = (fireIncidents["neighborhood"].value_counts())
fig, ax = plt.subplots(figsize=(20, 6))
colors = ['blue', 'green', 'red', 'orange']
labels.plot.bar(width = 0.5, ax=ax, color=colors)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Number of Fires by Neighborhood')
plt.show()
```

### Conclusion

After looking through the data, the metric can be applied to determine that Mt. Oliver Borough is the best neighborhood as it has the least amount of fires.
