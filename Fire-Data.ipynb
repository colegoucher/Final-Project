## Fire Data for neighborhoods in Pittsburgh

```python
import pandas as pd
import matplotlib.pyplot as plt

fireIncidents = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02")
pd.set_option('display.max_rows', None)
fireIncidents['neighborhood'].value_counts()
```

```python
bottom_10 = (fireIncidents['neighborhood'].value_counts()).iloc[:10]
plt.subplots(figsize=(7, 4))
bottom_10.plot.bar(color = 'red')
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Bottom 10 Neighborhoods by Number of Fires')
plt.show()
```

```python
top_10 = (fireIncidents['neighborhood'].value_counts()).iloc[-10:]
plt.subplots(figsize=(7, 4))
top_10.plot.bar(color = 'green').invert_xaxis()
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Top 10 Neighborhoods by Number of Fires')
plt.show()
```

```python
labels = (fireIncidents["neighborhood"].value_counts())
fig, ax = plt.subplots(figsize=(20, 6))
colors = ['blue', 'green', 'red', 'orange']
labels.plot.bar(width = 0.5, ax=ax, color=colors)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Number of Fires by Neighborhood')
plt.show()
```

Based on this data and the evaluating system, it can be concluded that Mt. Oliver Borough is the best neighborhood as it as the least amount of fires
