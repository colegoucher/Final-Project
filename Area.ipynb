### Area Data

Metric: A neighbourhood with more area would be more preferable

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
n=pd.read_csv("https://data.wprdc.org/datastore/dump/668d7238-cfd2-492e-b397-51a6e74182ff")
top10 = n.nlargest(100, 'sqmiles')
plt.figure(figsize=(20, 8))
plt.bar(top10['hood'], top10['sqmiles'])
plt.xlabel('Neighborhood')
plt.ylabel('Area (sqmiles)')
plt.title('Top 10 Neighborhoods by Area')
plt.xticks(rotation='vertical')
plt.show()
```
