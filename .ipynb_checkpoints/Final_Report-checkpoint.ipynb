### Kamal: Arrest
In order to find the best neighborhood we have to first take a look at neighborhoods with arrest rate. Doing this helps us decide which neighborhoods is safe to live and gets rid of the bad neighborhoods.

```python
import pandas as pd 
#import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Here the code reads the arrest file. 
#Then it will print out all the different neighborhoods 
#and the number of arrest that happenned there in decending order.
arrest = pd.read_csv("Arrest.csv")
pd.set_option('display.max_rows', None)
labels = (arrest["INCIDENTNEIGHBORHOOD"].value_counts())
print(labels)
```

```python
#Data set for the whole neighborhood in decending order from highest arrest rate to lowest. 
labels = (arrest["INCIDENTNEIGHBORHOOD"].value_counts())
# bar graph size expended so it does not overlap 
fig, ax = plt.subplots(figsize=(20, 6))
# color of the graphs
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
bar_width = .5
# labels for the graph 
labels.plot.bar(width = bar_width, ax=ax, color=colors)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Arrests')
plt.title('All the Neighborhoods by Number of Arrests ')
plt.show()
```

```python
#Data set for the top 10 most arrest neighborhood in decending order. 
# Here I'm just printing out the top 10 neighbiorhoods with the most number of arrest. 
top_10 = (arrest["INCIDENTNEIGHBORHOOD"].value_counts()).iloc[:10]
# the graph will not overlap so it expands it
fig, ax = plt.subplots(figsize=(20, 6))
# colors for the bar graph 
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
graph = top_10.plot.bar(color = colors)
# Adding labels
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Arrests')
plt.title('Top 10 Neighborhoods by Most Number of Arrests')
# Adding the number of arrests on top of each bar
for bar, value in zip(graph.patches, top_10):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05, str(value),
            ha='center', va='bottom', color='black')

plt.show()
```

```python
#Data set for the top 10 least arrested neighborhoods in decending order. 
# Here I'm just printing out the top 10 neighbiorhoods with the least number of arrest.
least_10 = (arrest["INCIDENTNEIGHBORHOOD"].value_counts()).iloc[-10:]
# the bar graph size so it is expanded 
fig, ax = plt.subplots(figsize=(20, 6))

# colors for the bar graph 
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
least_10.plot.bar(color=colors)
bars = ax.bar(least_10.index, least_10, color=colors)
# labels for the bar graph 
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Arrests')
plt.title('Top 10 Neighborhoods by Least Number of Arrests')
# Adding the number of arrests on top of each bar
for bar, value in zip(bars, least_10):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.001 * height, str(value),
            ha='center', va='bottom', color='black')

plt.show()
```

### Conclusion for Arrest dataset

After analysing the data MT.Oliver seems to be the best neighborhood with only two arrests. MT.Oliver tends to have the least amount of arrest which indicates less crime and so it is the best neighborhood for this dataset. 

### Cole: Fire 

The second metric in determining the best neighborhood in Pittsburgh was by the amount of fires each place had. A neighborhood that has less fires will be deemed as safer and thus a better neighborhood.

```python
import pandas as pd
import matplotlib.pyplot as plt
 
fireIncidents = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02")
pd.set_option('display.max_rows', None)
fireIncidents['neighborhood'].value_counts()
```

```python
bottom_10 = (fireIncidents['neighborhood'].value_counts()).iloc[:10]
plt.subplots(figsize=(7, 4))
bottom_10.plot.bar(color = 'red')
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Bottom 10 Neighborhoods by Number of Fires')
plt.show()
```

```python
top_10 = (fireIncidents['neighborhood'].value_counts()).iloc[-10:]
plt.subplots(figsize=(7, 4))
top_10.plot.bar(color = 'green').invert_xaxis()
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Top 10 Neighborhoods by Number of Fires')
plt.show()
```

```python
labels = (fireIncidents["neighborhood"].value_counts())
fig, ax = plt.subplots(figsize=(20, 6))
colors = ['blue', 'green', 'red', 'orange']
labels.plot.bar(width = 0.5, ax=ax, color=colors)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Fires')
plt.title('Number of Fires by Neighborhood')
plt.show()
```

### Conclusion

Based on the data, Mt. Oliver Borough would be considered the best overall neighborhood because it had the least amount of fires


### Siddharth: Area

In order to find the best neighborhood in Pittsburgh, we decided to use Area as the variable to tie in both the arrest and fire data. A neighborhood with high crime and fires but in a smaller area will be much more dangerous than an area with low crime and fires but in a larger area.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
n=pd.read_csv("https://data.wprdc.org/datastore/dump/668d7238-cfd2-492e-b397-51a6e74182ff")
for index, row in n.iterrows():
    neighborhood_name = row['hood']  
    area_in_miles = row['sqmiles']  
    print(neighborhood_name, area_in_miles ,"sq miles")
```

```python
top10 = n.nlargest(100, 'sqmiles')
plt.figure(figsize=(20, 8))
plt.bar(top10['hood'], top10['sqmiles'])
plt.xlabel('Neighborhood')
plt.ylabel('Area (sqmiles)')
plt.title('Neighborhoods by Area in sqmiles')
plt.xticks(rotation='vertical')
plt.show()
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
n=pd.read_csv("https://data.wprdc.org/datastore/dump/668d7238-cfd2-492e-b397-51a6e74182ff")
top10 = n.nlargest(10, 'sqmiles')
plt.bar(top10['hood'], top10['sqmiles'])
plt.xlabel('Neighborhood')
plt.ylabel('Area (sqmiles)')
plt.title('Top 10 Neighborhoods by Area')
plt.xticks(rotation='vertical')
plt.show()
```

```python
n=pd.read_csv("https://data.wprdc.org/datastore/dump/668d7238-cfd2-492e-b397-51a6e74182ff")
top10 = n.nsmallest(10, 'sqmiles')
plt.bar(top10['hood'], top10['sqmiles'])
plt.xlabel('Neighborhood')
plt.ylabel('Area (sqmiles)')
plt.title('Bottom 10 Neighborhoods by Area')
plt.xticks(rotation='vertical')
plt.show()
```

### Conclusion for Area Dataset

Based on the data being shown and our preference for larger neighborhoods, Squirrel Hill South would be the best neighbourhood according to this dataset.

### Danger Rating

Danger Rating is the metric that we have created to determine which neighborhood is overall the best based on a combination of our three datasets. \
The formula for Danger Rating is: \
        Danger Rating=Arrests+10*(Fires)/area \
We decided to multiply the number of fires by 10 since the number of fires is much smaller than the number of arrests. So if we did not multiply the fires by a big amount, Crimes would be the biggest factor upon which this scale would be decided upon. A neighborhood could have a low amount of crimes but a large amount of fires and still would be designated as safe if we did not multiply the number of fires. 

A low danger rating would mean a safer neighborhood and would be our pick for the best neighborhood. 
                      

```python
n=pd.read_csv("neighbourhood.csv")
top10 = n.nlargest(100, 'Danger rating')
plt.figure(figsize=(20, 8))
plt.bar(top10['Neighbourhood'], top10['Danger rating'])
plt.xlabel('Neighborhood')
plt.ylabel('Danger Rating')
plt.title('All neighborhoods according to our criteria')
plt.xticks(rotation='vertical')
plt.show()
```
