### Danger Rating

**Metric**: Arrests+10(Fires)/Area

```python
import os
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

n=pd.read_csv("neighbourhood.csv")
top10 = n.nlargest(100, 'Danger rating')
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
plt.figure(figsize=(20, 8))
plt.bar(top10['Neighbourhood'], top10['Danger rating'], color=colors)
plt.xlabel('Neighborhood')
plt.ylabel('Safety Rating')
plt.title('All the neighborhoods according to our criteria')
plt.xticks(rotation='vertical')
plt.show()
```

```python
n=pd.read_csv("neighbourhood.csv")
top10 = n.nlargest(10, 'Danger rating')
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
plt.bar(top10['Neighbourhood'], top10['Danger rating'], color=colors)
plt.xlabel('Neighborhood')
plt.ylabel('Safety Rating')
plt.title('Bottom 10 neighbourhoods according to our criteria')
plt.xticks(rotation='vertical')
plt.show()
```

```python
n=pd.read_csv("neighbourhood.csv")
top10 = n.nsmallest(10, 'Danger rating')
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
plt.bar(top10['Neighbourhood'], top10['Danger rating'], color=colors)
plt.xlabel('Neighborhood')
plt.ylabel('Safety Rating')
plt.title('Top 10 neighbourhoods according to our criteria')
plt.xticks(rotation='vertical')
plt.show()
```
