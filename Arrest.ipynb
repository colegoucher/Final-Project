# Arrest Data

```python
import pandas as pd 
#import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

**Metric**: Least number of arrest is consider to be the best neighborhood.

```python
# Here the code reads the arrest file. Then it will print out all the different neighborhoods and the number of arrest that happenned there in decending order.
arrest = pd.read_csv("Arrest.csv")
pd.set_option('display.max_rows', None)
labels = (arrest["INCIDENTNEIGHBORHOOD"].value_counts())
print(labels)
```

```python
#Data set for the whole neighborhood in decending order from highest arrest rate to lowest. 
labels = (arrest["INCIDENTNEIGHBORHOOD"].value_counts())
# bar graph size expended so it does not overlap 
fig, ax = plt.subplots(figsize=(20, 6))
# color of the graphs
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
bar_width = .5
# labels for the graph 
labels.plot.bar(width = bar_width, ax=ax, color=colors)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Arrests')
plt.title('All the Neighborhoods by Number of Arrests ')
plt.show()

```

```python
#Data set for the top 10 most arrest neighborhood in decending order. 
# Here I'm just printing out the top 10 neighbiorhoods with the most number of arrest. 
top_10 = (arrest["INCIDENTNEIGHBORHOOD"].value_counts()).iloc[:10]
# the graph will not overlap so it expands it
fig, ax = plt.subplots(figsize=(20, 6))
# colors for the bar graph 
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
graph = top_10.plot.bar(color = colors)
# Adding labels
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Arrests')
plt.title('Top 10 Neighborhoods by Most Number of Arrests')
# Adding the number of arrests on top of each bar
for bar, value in zip(graph.patches, top_10):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05, str(value),
            ha='center', va='bottom', color='black')

plt.show()

```

```python
#Data set for the top 10 least arrested neighborhoods in decending order. 
# Here I'm just printing out the top 10 neighbiorhoods with the least number of arrest.
least_10 = (arrest["INCIDENTNEIGHBORHOOD"].value_counts()).iloc[-10:]
# the bar graph size so it is expanded 
fig, ax = plt.subplots(figsize=(20, 6))

# colors for the bar graph 
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan']
least_10.plot.bar(color=colors)
bars = ax.bar(least_10.index, least_10, color=colors)
# labels for the bar graph 
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Arrests')
plt.title('Top 10 Neighborhoods by Least Number of Arrests')
# Adding the number of arrests on top of each bar
for bar, value in zip(bars, least_10):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.001 * height, str(value),
            ha='center', va='bottom', color='black')

plt.show()
```
